## **MICE**

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
imputer = IterativeImputer(estimator=dtc)
```

```python
ladata.columns
```

```python
ladata.isna().sum()
```

```python
ladata.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
ladata["Education"] = LabelEncoder().fit_transform(ladata["Education"])
ladata["Marital Status"] = LabelEncoder().fit_transform(ladata["Marital Status"])
ladata = ladata.drop(["Date"], axis=1)
```

```python
ladata.head()
```

```python
imputed = imputer.fit_transform(ladata)
```

## **DBSCAN Clustering**

```python
from sklearn.cluster import DBSCAN
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]

for i in range(1,20):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(eldata)
  wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,20), wcss)
```
